```
An attribute is a property of an instance that can change while the category remains the same. Example: a vehicle being parked/stopped/moving, and whether or not a bicycle has a rider.
```

---
- > **From sample_annotation:**
        
    **translation**: La ubicación del cuadro delimitador en metros ($x, y, z$). Fundamental para la posición actual y para calcular la velocidad histórica.

    **rotation**: La orientación como cuaternión ($w, x, y, z$). Necesario para calcular el heading (yaw) y el seno/coseno del ángulo.

    **size**: El tamaño en metros (ancho, largo, alto). Ayuda al modelo a distinguir entre un peatón (pequeño) y un autobús (grande) para evitar colisiones.

    **instance_token**: Clave foránea para identificar qué instancia de objeto es. Crucial para agrupar anotaciones a lo largo del tiempo y crear la historia del agente.   

    **num_lidar_pts / num_radar_pts**: Número de puntos lidar/radar en la caja. Útil para filtrar: si un objeto tiene 0 o muy pocos puntos, quizás sea un "falso positivo" o esté muy lejos, y es mejor no entrenar con él.
---

- > **From category:**

    **name**: El nombre de la categoría (ej. vehicle, human.pedestrian). Esto se suele convertir en un vector one-hot (embedding de tipo) para alimentar al modelo.
---

- > **From ego_pose:**
    
    **translation**: Origen del sistema de coordenadas en metros.

    **rotation**: Orientación del sistema de coordenadas.

    Nota: Se usan para transformar los datos de sample_annotation (que están en coordenadas globales) al marco de referencia del vehículo en ese instante.
---

- > **From log:**

    **location**: El área donde se capturó el registro (ej. singapore-onenorth). Es indispensable para que la API NuScenesMap sepa qué mapa cargar (Boston vs Singapur).
---

- > **From sample**:

    **timestamp**: Marca de tiempo Unix. Necesario para calcular velocidades y aceleraciones reales ($\Delta distancia / \Delta tiempo$).

    **prev / next**: Claves foráneas al sample anterior y siguiente. Permiten recorrer la secuencia temporalmente.
---

- > **From visibility:**

    **level**: Nivel de visibilidad (ej. 60-80%).

    **Utilidad**: Deberías considerar ignorar o dar menos peso en la loss function a objetos con visibilidad muy baja (0-40%), ya que su posición etiquetada podría ser inexacta o "adivinada".
